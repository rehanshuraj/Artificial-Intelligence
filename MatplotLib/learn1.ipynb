```python
import matplotlib.pyplot as plt

x=[1, 2, 3, 4, 5]
y=[10,20,15,25,30]

# plt draws line while connecting x,y points
plt.plot(x,y)

# to display on screen use show function
plt.show()
```

```python
x = ['Mon','Tues','Wed','Thur','Fri'] #x-axis
y = [10,15,7,20,12] #y-axis
#plt.plot(x,y, color='color name', linestyle='line_style',linewidth=value,marker='marker symbol', label='label name')
plt.plot(x,y)
plt.title('Bakery Sales This Week!')
plt.xlabel('Day of the week')
plt.ylabel('Sales Per Day')
plt.show()
```

```python
months = [1,2,3,4]
sales = [1000,1500,1200,1800]

plt.plot(months, sales, color='red', linestyle='--', linewidth = 2, marker= 'o', label='2025 sales data')
plt.xlabel('months')
plt.ylabel('sales per months')
plt.title('Monthly sales Data Report')
plt.grid(color='gray', linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
# xticks is used for replacing months , sales
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.legend()
plt.show()
```

# Bar Charts
### taller the bar , largest the sales

```python
#plt.bar(x,height,color= 'color name', width= value, label='label name')

product = ['A','B','C','D']
sales = [1000,1500,800,1200]

plt.bar(product,sales,color='orange',label='Sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('product Sale Comparison')
plt.legend()
plt.show()
```

# Pie-Chart
### pie-chart represents values in percentage

```python
#plt.pie(Values, labels=label_list, colors=color_list, autopct='%1.1f%%)

regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title("Revenue Contributon By Region")
plt.show()
```

# Histogram 
### in histogram values distributed in range


```python
#plt.hist(data, bins=number_of_bins, color='colorname', edgeecolor='black')

scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

plt.hist(scores, bins=5, color='purple', edgecolor='black')
plt.xlabel('Number of Students')
plt.ylabel('Score Distribution of Students')
# Height of bar = No. of Student in that range
plt.show()
```

# Scatter plot


```python
#plt.scatter(x,y,color='color_name', marker='marker_style', label='label_ name')

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

plt.scatter(hours_studies,exam_scores, color='green', marker='o', label='Student Data')
plt.xlabel('hours_studies')
plt.ylabel('exam_scores')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3], [50,55,60], color='blue', label='Class A') #g1
plt.scatter([1,2,3], [45,50,52], color='orange', label='Class B') #g1
plt.xlabel('hours_studies')
plt.ylabel('exam_scores')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

# SubPlot


```python
#plt.subplot(nrows, ncols, index)

x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) #1 row , 2 column, 1st subPlot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2) #1 row , 2 column, 2st subPlot
plt.bar(x,y)
plt.title('Bar chart')

plt.tight_layout()
plt.show()
```
